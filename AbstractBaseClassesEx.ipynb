```python
import abc 

class Scanner(abc.ABC):
    serialNumber = 0
    def __init__(self):
        Scanner.serialNumber +=  1

    @abc.abstractmethod
    def scan_document(self):
        return f"The document has been scanned"
    
    @abc.abstractmethod
    def get_scanner_status(self, max, resolution, serialNumber):
        return f"{max} - {resolution} - {serialNumber}"
    
class Printer(abc.ABC):
    serialNumber = 0
    def __init__(self):
        Printer.serialNumber += 1

    @abc.abstractmethod
    def print_document(self):
        return f"The document has been printed"
    
    @abc.abstractmethod
    def get_printer_status(self, max, resolution, serialNumber):
        return f"{max} - {resolution} - {serialNumber}"
```

```python
class MFD1(Scanner, Printer):
    resolution = "Low"
    max = "32"
    def __init__(self):
        self.serialNumber = f"P{Printer.serialNumber}-S{Scanner.serialNumber}"

    def scan_document(self):
        return super().scan_document()
    
    def get_scanner_status(self):
        return (max, MFD1.max, MFD1.resolution, self.serialNumber)
    
    def print_document(self):
        return super().print_document()
    
    def get_printer_status(self):
        return (MFD1.max, MFD1.resolution, self.serialNumber)


class MFD2(Scanner, Printer):
    resolution = "Medium"
    max = "64"
    def __init__(self):
        self.serialNumber = f"P{Printer.serialNumber}-S{Scanner.serialNumber}"
    
    def scan_document(self):
        return super().scan_document()
    
    def get_scanner_status(self):
        return (max, MFD2.max, MFD2.resolution, self.serialNumber)
    

    def print_document(self):
        return super().print_document()()
    
    def get_printer_status(self):
        return (MFD2.max, MFD2.resolution, self.serialNumber)
    
class MFD3(Scanner, Printer):
    resolution = "High"
    max = "124"
    def __init__(self):
        self.serialNumber = f"P{Printer.serialNumber}-S{Scanner.serialNumber}"
    
    def scan_document(self):
        return super().scan_document()
    
    def get_scanner_status(self):
        return (max, MFD3.max, MFD3.resolution, self.serialNumber)
    

    def print_document(self):
        return super().print_document()()
    
    def get_printer_status(self):
        return (MFD3.max, MFD3.resolution, self.serialNumber)
```

```python
mfd1 = MFD1()
mfd2 = MFD2()
mfd3 = MFD3()
```
